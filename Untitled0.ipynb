```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow_addons as tfa
import matplotlib.pyplot as plt
```

```python
#Cifar Dataset indirildi.
num_classes = 10
input_shape = (32, 32, 3)
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
print(f"x_train shape: {x_train.shape} - y_train shape: {y_train.shape}")
print(f"x_test shape: {x_test.shape} - y_test shape: {y_test.shape}")
```

```python
learning_rate = 0.001
weight_decay = 0.0001
batch_size = 256 # Bir seferde işlenen eğitim verisi örneklerinin sayısını belirler.
num_epochs = 40 # Eğitim süresince kaç epoch yapılacağını belirler.
image_size = 72  # Giriş görüntülerinin boyutunu belirler
patch_size = 6  # Giriş görüntülerinden çıkarılacak olan yama (patch) boyutunu belirler.
num_patches = (image_size // patch_size) ** 2
projection_dim = 64
num_heads = 4
transformer_units = [
    projection_dim * 2,
    projection_dim,
]
transformer_layers = 8 #Toplamda kaç adet transformer katmanının kullanılacağını belirler.
mlp_head_units = [2048, 1024]  # Son sınıflandırıcı katmanındaki dense katmanların boyutlarını belirler.
```

```python
# augmentation(veri artırma) model oluşturma
data_augmentation = keras.Sequential(
    [
        layers.Normalization(),
        layers.Resizing(image_size, image_size),
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(factor=0.02),
        layers.RandomZoom(
            height_factor=0.2, width_factor=0.2
        ),
    ],
    name="data_augmentation",
)
# Normalizasyon katmanına,
# eğitim veri setinin istatistiksel özelliklerini (ortalama ve varyans) öğretmek için kullanılır.
data_augmentation.layers[0].adapt(x_train)
```

```python
# (MLP - Multilayer Perceptron) oluşturan bir fonksiyonu tanımlar.
def mlp(x, hidden_units, dropout_rate):
    for units in hidden_units:
        x = layers.Dense(units, activation=tf.nn.gelu)(x)
        x = layers.Dropout(dropout_rate)(x)
    return x
```

```python
#Patch(yama) Çıkarma İşlemi
class Patches(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size


    def call(self, images):
        batch_size = tf.shape(images)[0] # Giriş görüntülerinin batch boyutunu alır.
        patches = tf.image.extract_patches( # görüntülerden yamalar çıkarılır.
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding="VALID",
        )
        patch_dims = patches.shape[-1] #  Çıkarılan yamaların boyutunu alır.
        patches = tf.reshape(patches, [batch_size, -1, patch_dims]) #yamaların boyutunu düzenler ve yeniden şekillendirir
        return patches
```

```python
# Embedding işlemi gerçekleştirilir.
class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim) # Bir Dense katmanı oluşturur.
        self.position_embedding = layers.Embedding( # Bir embedding katmanı oluşturur.
            input_dim=num_patches, output_dim=projection_dim
        )


    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded
```

```python
#Vision Transformer (ViT) modelini oluşturma.
def create_vit_classifier():
    inputs = layers.Input(shape=input_shape)
    augmented = data_augmentation(inputs)
    patches = Patches(patch_size)(augmented)
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)


    # Transformer bloğunda birden fazla katmanı oluşturur.
    for _ in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)#1. Normalizasyon katmanı ekler.
        # MultiHeadAttention katmanı ile attention mekanizması uygulanır
        attention_output = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=projection_dim, dropout=0.1
        )(x1, x1)
        x2 = layers.Add()([attention_output, encoded_patches])#1. Skip connection eklenir.
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)#2. Normalizasyon katmanı ekler.
        x3 = mlp(x3, hidden_units=transformer_units, dropout_rate=0.1)# MLP UYGULANIR.
        encoded_patches = layers.Add()([x3, x2])#2. Skip connection eklenir.


    #[batch_size, projection_dim] boyutunda bir tensör oluşturur.
    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.Flatten()(representation)
    representation = layers.Dropout(0.5)(representation)
    #MLP uygulandı.
    features = mlp(representation, hidden_units=mlp_head_units, dropout_rate=0.5)
    #Çıkış katmanı olarak bir Dense katmanı eklenir.
    logits = layers.Dense(num_classes)(features)
    #model oluşturulur.
    model = keras.Model(inputs=inputs, outputs=logits)
    return model
```

```python
def run_experiment(model):
    optimizer = tfa.optimizers.AdamW(
        learning_rate=learning_rate, weight_decay=weight_decay
    )
    model.compile(
        optimizer=optimizer,
        loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
        metrics=[
            keras.metrics.SparseCategoricalAccuracy(name="accuracy"),
            keras.metrics.SparseTopKCategoricalAccuracy(5, name="top-5-accuracy"),
        ],
    )
    checkpoint_filepath = "/tmp/checkpoint"
    checkpoint_callback = keras.callbacks.ModelCheckpoint(
        checkpoint_filepath,
        monitor="val_accuracy",
        save_best_only=True,
        save_weights_only=True,
    )
    model.summary() # Modelin özetini ekrana yazdırır.
    history = model.fit(
        x=x_train,
        y=y_train,
        batch_size=batch_size,
        epochs=num_epochs,
        validation_split=0.1,
        callbacks=[checkpoint_callback],
    )
    model.load_weights(checkpoint_filepath) #En iyi performans gösteren ağırlıkları yükler.
    _, accuracy, top_5_accuracy = model.evaluate(x_test, y_test) #Modelin test verileri üzerinde performansını değerlendirir
    print(f"Test accuracy: {round(accuracy * 100, 2)}%")
    print(f"Test top 5 accuracy: {round(top_5_accuracy * 100, 2)}%")
    return history
```

```python
vit_classifier = create_vit_classifier()
```

```python
history = run_experiment(vit_classifier)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10,10))
axs[0].plot(history.history['loss'])
axs[0].plot(history.history['val_loss'])
axs[0].title.set_text('Eğitim Kaybı vs Doğrulama Kaybı')
axs[0].set_xlabel('Epochs')
axs[0].set_ylabel('Loss')
axs[0].legend(['Train', 'Val'])

axs[1].plot(history.history['accuracy'])
axs[1].plot(history.history['val_accuracy'])
axs[1].title.set_text('Eğitim Doğruluğu vs Doğrulama Doğruluğu')
axs[1].set_xlabel('Epochs')
axs[1].set_ylabel('Accuracy')
axs[1].legend(['Train', 'Val'])
plt.savefig('egitim_grafikleri.png')
plt.show()
```
